```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cleaned_data.csv")
```

```python
data
```

```python
data_pivot = data.pivot_table(values = 'quantity',index='invoice',columns = 'stockcode',aggfunc = np.sum).fillna(0)
```

```python
data_pivot['Sum'] = data_pivot.sum(axis = 1)
data_pivot
```

```python
data_pivot.columns
```

```python
for name in data_pivot.columns:
    for i in range(len(data_pivot)):
        if data_pivot.loc[i,name]!= 0:
            data_pivot.loc[i,name] = name
    
```
